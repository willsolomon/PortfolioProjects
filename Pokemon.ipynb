```python
poke = poke[poke['Name'].str.contains('Mega')==False]
```

**Boxplot of Total States per Generation**

```python
totGen = sns.boxplot(data=poke, x = 'Generation', y='Total',palette='prism')
totGen.set(title='Total Stats per Generation')
```

**Distribution of Total Stats**

```python
totHist = sns.histplot(data=poke, x='Total')
#Eternatus Eternamax skewing data right. Bimodal w/peaks at ~300 and ~480
```

**Drop EternatusEternamax and Boolean Columns to Work With Pairplot**

```python
poke = poke[poke['Name'].str.contains('Eternamax')==False]
```

```python
poke = poke.drop('Legendary',axis=1)
poke = poke.drop('Gigantimax',axis=1)
```

```python
poke.head()
```

**Resetting Index**

```python
poke.reset_index(inplace=True)
```

```python
poke.head()
```

**Seaborn Plots to Visually Analyze Data**

```python
pokePair = sns.pairplot(data=poke)
```

```python
pokeCorr = poke.corr()
```

```python
pHeatMap = sns.heatmap(pokeCorr,linewidths = .04, linecolor = 'black', annot=True)
pHeatMap.set(xlabel='Pokemon Attributes',ylabel = 'Pokemon Attributes',title = 'Correlation of Pokemon Attributes')
```

**Visualizing Possible Correlations**

```python
'''Possible Correlations (above .4 and excluding Total):
    HP - Attack
    Attack - Defense
     Defense - Sp. Def
    Sp. Atk - Sp. Def
    Sp. Atk - Speed
    '''
fig, axes = plt.subplots(2,3, figsize = (18,12))
sns.regplot(ax = axes[0,0],data=poke, x = 'HP',y='Attack',scatter_kws={'color': 'blue'},line_kws={'color':'yellow'})
axes[0,0].set_title('HP vs. Attack')

sns.regplot(ax = axes[0,1],data=poke, x = 'Attack',y='Defense',scatter_kws={'color': 'blue'},line_kws={'color':'yellow'})
axes[0,1].set_title('Attack vs. Defense')

sns.regplot(ax = axes[0,2],data=poke, x = 'Defense',y='Sp. Def',scatter_kws={'color': 'blue'},line_kws={'color':'yellow'})
axes[0,2].set_title('Defense vs. Sp. Def')

sns.regplot(ax = axes[1,0],data=poke, x = 'Sp. Atk',y='Sp. Def',scatter_kws={'color': 'blue'},line_kws={'color':'yellow'})
axes[1,0].set_title('Sp. Atk vs. Sp. Def')

sns.regplot(ax = axes[1,1],data=poke, x = 'Sp. Atk',y='Speed',scatter_kws={'color': 'blue'},line_kws={'color':'yellow'})
axes[1,1].set_title('Sp. Atk vs. Speed')

fig.delaxes(axes[1,2])
```
